# C# in Jupyter Notebooks — Proof of Concept
## .NET Interactive Kernel Tutorial

This notebook demonstrates that C# runs successfully in Jupyter via the `.NET Interactive` kernel.
It covers basic C# features, LINQ, async patterns, and data structures.

**Kernel:** .NET (C#) via `dotnet-interactive`

---

## 1. Hello World — Basic C# in Jupyter

```
// Basic output — no Main() method needed in .NET Interactive!
Console.WriteLine("Hello from C# in Jupyter!");
Console.WriteLine($"Running on .NET {Environment.Version}");
Console.WriteLine($"OS: {Environment.OSVersion}");
```

## 2. Variables and String Interpolation

```
// Variables and string interpolation
string name = "Alex";
int year = DateTime.Now.Year;
double pi = Math.PI;

Console.WriteLine($"Hello {name}! Welcome to {year}.");
Console.WriteLine($"Pi to 6 decimal places: {pi:F6}");

// Type inference with var
var languages = new[] { "C#", "Python", "Julia", "F#" };
Console.WriteLine($"Languages available in Jupyter: {string.Join(", ", languages)}");
```

## 3. LINQ — Language Integrated Query

```
using System.Linq;

// Generate sample data
var numbers = Enumerable.Range(1, 20);

// LINQ queries
var evens = numbers.Where(n => n % 2 == 0);
var squares = numbers.Select(n => n * n);
var sumOfOdds = numbers.Where(n => n % 2 != 0).Sum();

Console.WriteLine($"Evens: {string.Join(", ", evens)}");
Console.WriteLine($"First 5 squares: {string.Join(", ", squares.Take(5))}");
Console.WriteLine($"Sum of odd numbers 1-20: {sumOfOdds}");

// LINQ with anonymous types
var stats = new[] { 85, 92, 78, 96, 88 }
    .Select((score, i) => new { Student = $"Student {i+1}", Score = score })
    .OrderByDescending(s => s.Score);

Console.WriteLine("\nStudent Rankings:");
foreach (var s in stats)
    Console.WriteLine($"  {s.Student}: {s.Score}");
```

## 4. Classes and Records

```
// Records — immutable data types (C# 9+)
record Person(string Name, int Age, string Role);

var team = new List<Person>
{
    new("Alice", 30, "Developer"),
    new("Bob", 28, "Designer"),
    new("Charlie", 35, "Architect"),
    new("Diana", 32, "DevOps")
};

// Pattern matching (C# 8+)
foreach (var person in team)
{
    var description = person switch
    {
        { Age: < 30 } => "Junior",
        { Role: "Architect" } => "Senior Technical Lead",
        { Age: >= 30 } => "Experienced",
        _ => "Unknown"
    };
    Console.WriteLine($"{person.Name} ({person.Age}): {description}");
}
```

## 5. Async/Await

```
using System.Threading.Tasks;

// Async methods work in .NET Interactive
async Task<string> FetchDataAsync(string source)
{
    await Task.Delay(500); // Simulate network delay
    return $"Data from {source} (fetched at {DateTime.Now:HH:mm:ss})";
}

// Parallel async execution
var tasks = new[] { "API-1", "API-2", "API-3" }
    .Select(source => FetchDataAsync(source));

var results = await Task.WhenAll(tasks);

Console.WriteLine("Parallel fetch results:");
foreach (var result in results)
    Console.WriteLine($"  {result}");
```

## 6. Collections and Dictionaries

```
// Dictionary with collection initializer
var inventory = new Dictionary<string, (int Quantity, decimal Price)>
{
    ["Widget A"] = (150, 9.99m),
    ["Widget B"] = (75, 24.99m),
    ["Gadget X"] = (200, 14.99m),
    ["Gadget Y"] = (30, 49.99m)
};

// LINQ aggregation
var totalValue = inventory.Sum(kv => kv.Value.Quantity * kv.Value.Price);
var mostExpensive = inventory.MaxBy(kv => kv.Value.Price);

Console.WriteLine("Inventory Report:");
foreach (var (name, (qty, price)) in inventory)
    Console.WriteLine($"  {name}: {qty} units @ ${price} = ${qty * price:N2}");

Console.WriteLine($"\nTotal Inventory Value: ${totalValue:N2}");
Console.WriteLine($"Most Expensive: {mostExpensive.Key} at ${mostExpensive.Value.Price}");
```

## 7. NuGet Package Installation

`.NET Interactive` supports installing NuGet packages directly in notebooks:

```
// Install and use a NuGet package
#r "nuget: Humanizer, 2.14.1"

using Humanizer;

Console.WriteLine("Humanizer Examples:");
Console.WriteLine($"  1234567 → {1234567.ToWords()}");
Console.WriteLine($"  DateTime.Now → {DateTime.Now.Humanize()}");
Console.WriteLine($"  'some_variable_name' → {"some_variable_name".Humanize()}");
Console.WriteLine($"  TimeSpan 3hrs → {TimeSpan.FromHours(3).Humanize()}");
```
